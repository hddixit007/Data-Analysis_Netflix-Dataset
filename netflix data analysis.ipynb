## Q2.  In which year highest no of TV shows & movies are released? Show with bar graph

```python
data.head()
```

#### Now here it is asking the year which is not present in our dataset. So we have to create a new column in the dataset and extract the year from the Release_Date and put in year column
##### 1. To Change the Release_Date to proper date format we have to use pandas function "to_datetime".
##### 2. data[ ' date ' ] = pd. to_datetime(data [Release_Date] )
##### 3. Finally data [ date ].year.value_counts() will give the count of year of movie releases
 

```python
data['date'] = pd.to_datetime(data['Release_Date'])
```

```python
data.head()
```

```python
data.dtypes
```

## dt.year--> It will convert date to year


```python
data['date'].dt.year
```

```python
data['date'].dt.year.value_counts()    # It counts the occurence of all individual years in the columns


## Therefore we can say that in 2019 max no of movies are released
```

## Bar graph
